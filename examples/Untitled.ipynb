```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport

df = pd.DataFrame(
    np.random.rand(100, 5),
    columns=["a", "b", "c", "d", "e"]
)

```

```python
profile = ProfileReport(df, title='Pandas Profiling Report', explorative=True)
```

```python
profile.to_widgets()
```

```python
profile.to_notebook_iframe()
```

```python
json_data = profile.to_json()
```
